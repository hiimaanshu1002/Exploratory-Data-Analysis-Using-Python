**What is the average BasePay ?**

```python
#converting all the not provided to nans for type conversion, so that int/float methods can be applied.

for index, row in sal.iterrows():
    for col in sal.columns:
        if (row[col] == 'Not provided') or (row[col] == 'Not Provided') :
            sal.at[index, col] = np.NaN   
```

```python
sal = sal.astype({'BasePay': 'float64'})
```

```python
sal.mean(numeric_only = True)
```

```python
sum = 0
count = 0
for val in sal['BasePay']:
    if np.isnan(val) == False:
        sum += val
        count += 1
print(sum/count)
```

```python
print(sal.mean(numeric_only = True)['BasePay'])
```

**What are the top 5 most common jobs?**

```python
sal['JobTitle'].value_counts().head()
```

```python
unique = {}
for title in sal["JobTitle"]:
    if title in unique:
        unique[title] += 1
        continue
    else:
        unique[title] = 1
        
jobs = pd.Series()

for key,value in unique.items():
    jobs[key] = value

jobs.sort_values(ascending = False).head(5)
```

**How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)**

```python
unique = {}
for title in sal['JobTitle'].loc[sal['Year'] == 2013]:
    if title in unique:
        unique[title] += 1
        continue
    else:
        unique[title] = 1
        
jobs = pd.Series()

for key,value in unique.items():
    jobs[key] = value

jobs[jobs == 1]
```

**How many people have the word Chief in their job title? (This is pretty tricky)**

```python
sal = pd.read_csv('Salaries.csv')
unique = {}
for title in sal["JobTitle"]:
    if title in unique:
        unique[title] += 1
        continue
    else:
        unique[title] = 1

num_chiefs = 0 
for key in unique.keys():
    if 'Chief' in key:
        num_chiefs += unique[key]
print(num_chiefs)
```

**Is there a correlation between length of the Job Title string and Salary?**

```python
relation = pd.DataFrame(sal[['JobTitle','TotalPayBenefits']])
```

```python
relation['LengthOfJobTitle'] = [len(i) for i in sal['JobTitle']]
```

```python
relation.sort_values(by=['TotalPayBenefits','LengthOfJobTitle'], ascending=False)
```
